Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)"

```python
import pandas as pd
import numpy as np
```

```python
match='Коефіцієнт народжуваності в регіонах України'
table = pd.read_html('https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8',thousands="", decimal=",",
                          match = match,
                    )
table
```

* Вивести перші рядки таблиці за допомогою методу head

```python
table[0].head()
```

* Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
table[0].shape
```

```python
df = pd.DataFrame(table[0])
df.shape
```

* Замініть у таблиці значення "—" на значення NaN

```python
df = df.replace(to_replace=r'[^A-Za-z0-9\.А-Яа-яіїІ-]', value=np.nan, regex=True)
df
```

* Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
df.dtypes
```

* Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
df = df.astype({'2014': np.float64, '2019': np.float64})
df.dtypes
```

* Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
df_nan = pd.isnull(df).sum()
print(f'Частка пропусків в %:\n{df_nan*100/len(df)}')
```

* Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
df_drop = df.drop(index=27)
df_drop
```

* Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
df_means = df_drop.mean()
df_means
```

```python
df_without_nan = df_drop.fillna(df_means)#, inplace=True)
df_without_nan
```

* Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
df_means['2019']
```

```python
df_without_nan['Регіон'][df_without_nan['2019'] > (df_means['2019'])]
```

* У якому регіоні була найвища народжуваність у 2014 році?

```python
max_2014 = df_without_nan['2014'].max()
max_2014
```

```python
df_without_nan[df_without_nan['2014'] == max_2014]['Регіон'].values[0]
```

* Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
import matplotlib.pyplot as plt
```

```python
data_2019 = df_without_nan['2019']
data_region = df_without_nan['Регіон']
plt.bar(data_region, data_2019)
plt.xlabel("Країна", fontsize=6)
plt.xticks(rotation=90)
plt.ylabel("Кількість", fontsize="small")
plt.title("Народжуваність по регіонах у 2019 році", fontsize=15)
plt.show()
```

## Візуалізація
додайте від 3 до 5 графіків функцій різного типу на ваш вибір.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df_without_nan.head()
```

```python
# Коробковий графік може показати розподіл населення для кожного року та виявити викиди
data_population_by_years = df_without_nan.drop(columns=['Регіон'])
plt.figure(figsize=(10, 6))
plt.boxplot(data_population_by_years.values, labels=data_population_by_years.columns)
plt.title('Розподілу населення за роками')
plt.xlabel('Рік')
plt.ylabel('Населення')
plt.grid(True)
plt.show()
```

```python
data_set_idx_region = df_without_nan.set_index(['Регіон'])#, inplace=True)
```

```python
# Теплова карта може допомогти візуалізувати зміни у розмірі населення у всіх регіонах за всі роки, використовуючи кольорову шкалу
plt.figure(figsize=(10, 6))
sns.heatmap(data_set_idx_region, annot=True, fmt=".1f", cmap="coolwarm", linewidths=0.5)
plt.title('Теплова карта розподілу населення за роками та регіонами', fontsize=15)
plt.xlabel('Рік', fontsize=14)
plt.ylabel('Регіон', fontsize=14)
plt.show()
```

```python
# 3D-графік відображає динаміку населення в регіонах за кілька років.
regions = df_without_nan['Регіон']
years = df_without_nan.columns[1:].astype(int)
population = df_without_nan.iloc[:, 1:]

fig = plt.figure(figsize=(15, 12))
ax = fig.add_subplot(111, projection='3d')

# Побудова точок для кожного регіону та кожного року
for i in range(len(regions)):
    xs = years
    ys = [i] * len(years)
    zs = population.iloc[i]
    ax.plot(xs, ys, zs)

ax.set_yticks(range(len(regions)))
ax.set_yticklabels(regions, rotation=90, fontsize=8)

ax.set_xlabel('Рік')
ax.set_ylabel('Регіон', labelpad=24)
ax.set_zlabel('Населення')

plt.title('3D-графік розподілу населення за роками та регіонами')
plt.show()
```
