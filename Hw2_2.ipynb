* Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

* Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

```python
min_max = data_grouped.agg(['min', 'max'])
min_max
```

```python
data_salary = min_max['Зарплата на місяць']
data_salary
```

* Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
def fill_avg_salary(row):
    return row.mean()

data_salary_min_max_avg = data_salary.copy()
data_salary_min_max_avg.loc[:, 'avg'] = data_salary_min_max_avg.apply(fill_avg_salary, axis=1)
data_salary_min_max_avg
```

```python
data_salary_min_max_avg['avg'].describe()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

* Кругова діаграма: Розподіл кількості респондентів за мовами програмування

```python
# Підрахунок кількості респондентів для кожної мови програмування
language_counts = statistic_cleaned['Мова програмування'].value_counts()
other_lang = language_counts.loc[language_counts <= language_counts['Other']].sum()
data_language = pd.concat([language_counts.loc[language_counts > language_counts['Other']], pd.Series(data=[other_lang], index=["Other"])])

# Побудова кругової діаграми
plt.figure(figsize=(10, 6))
data_language.plot(kind='pie', autopct='%1.1f%%', shadow=True, pctdistance=0.9, labeldistance=1.1) #, startangle=140)
plt.title('Розподіл кількості респондентів за мовами програмування')
plt.ylabel(None)
plt.axis('equal')  # Забезпечує круглу форму діаграми
plt.show()
```

* Гістограма: Розподіл зарплат за досвідом роботи

```python
# заміна типу не числових рядків на числові
pd = python_data.replace({'Загальний досвід роботи': {'10 і більше років': 10, 'менше 3 місяців': 0.3}})
pd = pd.astype({'Загальний досвід роботи': np.float64})
```

```python
# Розділення даних за категоріями досвіду роботи
low_exp = pd[pd['Загальний досвід роботи'] <= 1]
medium_exp = pd[(pd['Загальний досвід роботи'] > 1) & (pd['Загальний досвід роботи'] <= 5)]
high_exp = pd[pd['Загальний досвід роботи'] > 5]
```

```python
# Побудова гістограми для кожної категорії досвіду роботи
plt.figure(figsize=(10, 6))
plt.hist(low_exp['Зарплата на місяць'], bins=10, color='skyblue', alpha=0.5, label='Низький досвід')
plt.hist(medium_exp['Зарплата на місяць'], bins=10, color='salmon', alpha=0.5, label='Середній досвід')
plt.hist(high_exp['Зарплата на місяць'], bins=10, color='green', alpha=0.5, label='Високий досвід')

plt.legend()
plt.title('Розподіл зарплат за досвідом роботи (Python)')
plt.xlabel('Зарплата на місяць')
plt.ylabel('Частота')
plt.grid(True)
plt.show()
```

* Графік плотності: Порівняння розподілу зарплат для різних мов програмування

```python
# Побудова графіка плотності
plt.figure(figsize=(10, 6))
sns.kdeplot(data=statistic_cleaned, x='Зарплата на місяць', hue='Мова програмування', fill=True, warn_singular=False)
plt.title('Порівняння розподілу зарплат для різних мов програмування')
plt.xlabel('Зарплата на місяць')
plt.ylabel('Щільність')
plt.grid(True)
plt.show()
```
