* Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
statistic_it_vacancies_na = pd.isnull(statistic_it_vacancies).sum()
persentege_of_missing_data = statistic_it_vacancies_na*100/len(statistic_it_vacancies)
print(f'Частка пропусків в %:\n{persentege_of_missing_data}')
```

* Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"

```python
missing_data = persentege_of_missing_data[persentege_of_missing_data > 0]
columns_to_drop = missing_data.index.drop(labels='Мова програмування')
columns_to_drop
```

```python
statistic_cleaned = statistic_it_vacancies.drop(columns=columns_to_drop)
statistic_cleaned
```

* Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся, що залишився тільки стовпець "Мова.програмування"

```python
statistic_it_vacancies_na = pd.isnull(statistic_cleaned).sum()
persentege_of_missing_data = statistic_it_vacancies_na*100/len(statistic_cleaned)
print(f'Частка пропусків в %:\n{persentege_of_missing_data}')
```

* Видаліть усі рядки у вихідній таблиці за допомогою методу dropna

```python
statistic_cleaned = statistic_cleaned.dropna()
statistic_cleaned
```

* Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python

```python
python_data = statistic_cleaned.loc[(statistic_cleaned['Мова програмування'] == 'Python')]
python_data
```

* Визначте розмір таблиці python_data за допомогою методу shape

```python
python_data.shape
```

* Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

```python
data_grouped = python_data.groupby(['Посада'])
data_grouped
```

* Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

```python
min_max = data_grouped.agg(['min', 'max'])
min_max
```

```python
data_salary = min_max['Зарплата на місяць']
data_salary
```

* Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
def fill_avg_salary(row):
    return row.mean()

data_salary.loc[:]['avg'] = data_salary.apply(fill_avg_salary, axis=1)
data_salary
```

* Створіть описову статистику за допомогою методу describe для нового стовпчика.

```python
data_salary['avg'].describe()
```
